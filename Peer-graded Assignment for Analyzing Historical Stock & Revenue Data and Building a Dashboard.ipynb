```python
import yfinance as yf

# Define the ticker symbol
tickerSymbol = 'TSLA'

# Get data for the stock
tesla_data = yf.Ticker(tickerSymbol).history(period='1d', start='2020-01-01', end='2021-12-31')

# Reset the index
tesla_data = tesla_data.reset_index()

# Save the data to a CSV file
tesla_data.to_csv('tesla_data.csv', index=False)

# Display the first five rows of the dataframe
print(tesla_data.head())
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Make a GET request to the Tesla revenue website
url = 'https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue'
html_content = requests.get(url).text

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(html_content, 'html.parser')

# Find the revenue table in the HTML content
revenue_table = soup.find('table', {'class': 'table'})

# Extract the revenue data from the table and store it in a pandas dataframe
tesla_revenue = pd.read_html(str(revenue_table), header=0)[0]

# Display the last five rows of the dataframe
print(tesla_revenue.tail())
```

```python
import yfinance as yf
import pandas as pd

# Use yfinance to get Gamestop stock data from 2020-01-01 to 2021-01-01
gme = yf.download('GME', start='2020-01-01', end='2021-01-01')

# Reset the index
gme_data = gme.reset_index()

# Save the dataframe
gme_data.to_csv('gme_data.csv', index=False)

# Display the first five rows of the dataframe
print(gme_data.head())
```

```python
# Import necessary libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Define the URL to scrape
url = 'https://www.macrotrends.net/stocks/charts/GME/gamestop/revenue'

# Send a GET request to the URL and store the response object
res = requests.get(url)

# Use BeautifulSoup to parse the HTML content of the response
soup = BeautifulSoup(res.content,'html.parser')

# Find the revenue table in the HTML content
revenue_table = soup.find('table', {'class': 'historical_data_table'})

# Extract the revenue data from the table and store it in a pandas dataframe
gme_revenue = pd.read_html(str(revenue_table), header=0)[0]

# Display the last five rows of the dataframe
print(gme_revenue.tail())
```

```python
import matplotlib.pyplot as plt

# Define make_graph function
def make_graph(df, title):
    plt.plot(df.index, df['Close'])
    plt.xlabel('Date')
    plt.ylabel('Stock Price ($)')
    plt.title(title)
    plt.show()

# Plot Tesla stock graph
make_graph(tesla_data, 'Tesla Stock Price 2020-2021')
```

```python
import matplotlib.pyplot as plt

# Plot GameStop stock graph
make_graph(gme_data, 'GameStop Stock Price 2020-2021')

# Show the plot
plt.show()
```
